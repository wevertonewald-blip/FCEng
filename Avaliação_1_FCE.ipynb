```python
from graphviz import Digraph

g = Digraph()

g.node('A', 'Início')
g.node('B', 'C = 0')
g.node('C', 'Informar V atual')

g.node('D', 'V < 20?')
g.node('E', 'V < 80?')

g.node('F', 'Nível baixo')
g.node('G', 'Operação normal')
g.node('H', 'Risco de transbordar')

g.node('I', 'C = C + 1')
g.node('J', 'C = 3')

g.node('K', 'Alerta')
g.node('L', 'Normal')

g.node('N', 'Fim')

g.edge('A', 'B')
g.edge('B', 'C')
g.edge('C', 'D')

# V < 20
g.edge('D', 'F', label='Sim')
g.edge('D', 'E', label='Não')

# nível baixo
g.edge('F', 'I')
g.edge('I', 'J')

g.edge('J', 'K', label='Sim')
g.edge('J', 'L', label='Não')

g.edge('K', 'N')
# alerta indo para o fim
g.edge('L', 'C')
# normal indo para o inicio

# V < 80
g.edge('E', 'G', label='Sim')
g.edge('E', 'H', label='Não')

#operação normal e risco indo para C = 0
g.edge('G', 'B')
g.edge('H', 'B')

g
```

**Durante a operação de um processo, a concentração de um reagente é monitorada periodicamente para avaliar as condições de operação do sistema. A concentração pode variar ao longo do tempo e deve ser classificada em diferentes faixas para auxiliar na análise do processo. Neste contexto, você recebeu a tarefa de criar um algoritmo que permita executar
a simulação do sistema de monitoramento, o qual deve seguir os seguintes requisitos:**

**• A cada leitura, deve ser gerado um valor aleatório de concentração entre 0 e 10 mol/L.**

**• Se a concentração for menor que 3, o sistema deve indicar “Baixa”.**

**• Se a concentração estiver entre 3 e 7, o sistema deve indicar “Adequada”.**

**• Se a concentração for maior que 7, o sistema deve indicar “Alta”.**

**• Ao final da simulação, deve-se contabilizar quantas vezes cada condição ocorreu.**

**Com base nesses requisitos, desenvolva um código modularizado em Python para simular o
monitoramento considerando 10 leituras consecutivas.**

**O código deve conter, obrigatoriamente:**

**• Pelo menos uma função**

**• Estrutura de repetição**

**• Estrutura condicional**

**• Comentários explicativos**

**Ao final da execução, o programa deve apresentar:**

**• O número de ocorrências de cada classe (Baixa, Adequada e Alta)**

**• Uma mensagem indicando qual condição foi predominante no processo**

**Um exemplo de informações ao usuário pode ser visto no Quadro 1.**

image.png

```python
import numpy as np

#definição de constantes inicialmente
n=10
i=0
alta, baixa, adequada=0,0,0

#gerar 10 numeros aleatórios
for i in range(n):
  #necessita ter esse *n para gerar numeros até o valor 10
  C=np.random.rand()*n

  #verificando em que faixa cada valor se encontra
  if C<3:
    m='Baixa'
    baixa+=1
  elif C<=7:
    m='Adequada'
    adequada+=1
  else:
    m='Alta'
    alta+=1

  print(f'Leitura {i+1}: {C:.3f} mol/L se encontra em {m} concentração')

#identifica a quantidade de concentrações se enquadrando nas classes
print(f'\nLeituras em Baixa concentração: {baixa}')
print(f'Leitura em Adequada concentração: {adequada}')
print(f'Leitura em Alta concentração: {alta}')

#formação de um lista em que mostra a quantidade que as classes apareceram
predominante=[alta, baixa, adequada]
maiorqnt=max(predominante)

#verificando qual é a mais predominante
if predominante == baixa:
    print("\nCondição predominante: Baixa")
elif predominante == adequada:
    print("\nCondição predominante: Adequada")
else:
    print("\nCondição predominante: Alta")




```
