```python
import numpy as np

tolerancia = 1e-6
s=[]
N = 0

def termo_de_S(n): # mudei o nome da função
  return (-1)**n/(2*n+1)

for n in range(0,500):
  s.append(termo_de_S(n))
  soma = np.sum(s)

  if abs(soma-np.pi/4)**2 > tolerancia:
    N = N + 1
    mensagem = f"foram feitas {N+1} interações."

  else:
    mensagem = f'o valor fornecido de {N+1} já atende a função.'
    break

print(mensagem)
```
