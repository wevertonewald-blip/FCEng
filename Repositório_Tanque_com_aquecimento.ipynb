```python
from dataclasses import dataclass

@dataclass
class param:
  '''
m = massa do líq (kg)
cp = calor específico (J/kg .°C)
Q = potência de aquecimento (W)
U = Coeficiente globall de troca térmica(W/m².°C)
T = Temperatura (°C)
  '''

A: float = 5
cp: float = 1670
U: float = 150
T: int = 25
```

```python
def aqueciele(t, Tamb, paran, Q, m):
  dTdt = (Q-paran.U*paran.A*(paran.T-Tamb))/(m*paran.cp)
  return dTdt
```

```python
from scipy.integrate import solve_ivp
def simula(t, paran):
  sol = solve_ivp(aqueciele, t, [paran.T], args=(paran), method='RK45')
  return sol.t, sol.y[0]
```
