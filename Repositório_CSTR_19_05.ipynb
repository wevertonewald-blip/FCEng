Repositório para o exemplo da aula do dia 19/05, onde consite em um sistema com 3 reatores CSTR em série com determinados parametros

```python
from dataclasses import dataclass

@dataclass
class param:
  '''
  V: Volume em L
  F: Taxa da vazão L/min
  K: Constante de velocidade min^-1

  '''
  F: float
  V: float
  K: float
```

```python
def Model_reator(Cao, param):
  Cai=(param.F/(param.V+param.K))*Cao
  return Cai
```

```python
def simula(n, Cao):
  para=param(V=100, F=50, K=0.5)
  Calist=[Cao]
  Xlist=[]
  for i in range(n):
    Ca=Model_reator(Calist[i], para)
    X=(Calist[0]-Ca)/Calist[0]
    Calist.append(Ca)
    Xlist.append(X)
  return Xlist, Calist
```

```python
simula(3, 1)
```
