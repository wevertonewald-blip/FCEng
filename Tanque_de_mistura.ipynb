```python
from dataclasses import dataclass

@dataclass
class correntes:
  '''
  Vazão: vazão em kg/h
  Fração: fração massica da corrente em kg/kg
  '''
  Vazão: float
  Fração: float
```

```python
def mist(c1:correntes, c2: correntes):
  '''
  calcule a vazão e a fração massica das misturas de duas correntes
  c1:corrente A
  C2:corrente B
  x:fração massica de sal na saida
  Ft:vazão total da mistura
  FaS:Fração de sal corrente A
  FbS:Fração de sal corrente B
  '''
  FaS= c1.Vazão*c1.Fração
  FbS= c2.Vazão*c2.Fração
  Ft=c1.Vazão+c2.Vazão
  x=(FaS+FbS)/Ft
  return Ft,x
```
